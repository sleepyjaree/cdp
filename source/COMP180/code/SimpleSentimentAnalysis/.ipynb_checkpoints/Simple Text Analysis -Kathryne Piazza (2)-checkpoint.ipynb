# Simple text analytics

We start with a compact version of the code we used to look at **D**ocument **F**requencies (DF). As always, we start with the necessary housekeeping, importing the necessary toolboxes and Twitter API credentials.

```python
import os
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import itertools
import collections
import datetime

import tweepy as tw
import nltk
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize, sent_tokenize
import re
import networkx

import math
from textblob import TextBlob as tb

from collections import defaultdict

import warnings
warnings.filterwarnings("ignore")

from IPython.display import display, Markdown, Image



# Cosmetics for later ....
sns.set(font_scale=1.05)
sns.set_style("whitegrid")

# stopwords
nltk.download('stopwords')
stopWords = set(stopwords.words('english'))

##new
newStopWords = ["like", "would", "dont", "get", "one", "every", "also", "yes", "cant", "hes", "need", "make", "want",
               "mean", "even", "said", "self", "im", "youre", "say", "really", "h", "undergoing", "th", "going", "w", "says",
               "never", "thats", "isnt" ,"someone", "still", "something", "back", "know", "think", "true", "way",
               "see", "thing", "oh", "whats", "b", "didnt", "allegedly"]
stopWords = stopWords.union(newStopWords)
##new

# My access keys are in ./credentials.py as variables:
# CONSUMER_KEY, CONSUMER_SECRET, ACCESS_TOKEN, and ACCESS_SECRET 
#
# Importing them in this manner will allow me to 
# use them as variables, below.
#
from credentials import * 

# authenticating with Twitter

authenticateMe = tw.OAuthHandler(CONSUMER_KEY, CONSUMER_SECRET)
authenticateMe.set_access_token(ACCESS_TOKEN, ACCESS_SECRET)
api = tw.API(authenticateMe, wait_on_rate_limit=True)
```

## Define the scope of our search

The variable ``searchFor`` is the search argument we are passing to Twitter via the interface we've established above. We can include standard Twitter nomenclature here, e.g., ``-filter:retweets``, to narrow our search as needed. 

```python
# Parameters
# what to search for?
# what language to conduct the search in?
# since when (date)?
# until when (date)?
# how many tweets to pull in?
# how many terms to plot?
#
################################################################################

SEARCH_TERM        = "abortion"
LANGUAGE           = "en"
SINCE_DATE         = '2016-01-01' # eventually need datetime()
UNTIL_DATE         = '2020-01-01' # formatted yyyy-mm-dd
TWEETS_TO_BRING_IN = 2000          # KEEP THIS UNDER 3000 PLEASE!
TOP_N_TERMS        = 50

################################################################################

# the code below can take a few seconds to run, depending
# on the number of tweets we specify as argument in 
# method .items(). Run this cell sparingly to save time.
#
searchFor = SEARCH_TERM + " -filter:retweets"

tweets = tw.Cursor(api.search,
                   q=searchFor,
                   lang=LANGUAGE,
                   since=SINCE_DATE,
                   until=UNTIL_DATE).items(TWEETS_TO_BRING_IN)

textFromTweets = [tweet.text for tweet in tweets]

# Plain vanilla function to remove URI/URLs from tweets
# using the re package for regular expressions
#


def remove_url(txt):
    """Remove URL by replacing it with an empty
    substring within a given string txt 

    Parameters
    ----------
    txt : string
        the string to remove URLs from.

    Returns
    -------
    The same txt string without any URLs.
    """

    txt = re.sub("@[^\s]+","",txt)
    
    return " ".join(re.sub("([^0-9A-Za-z \t])|(\w+:\/\/\S+)", "", txt).split())

##new
from nltk.stem import LancasterStemmer      #uses LancasterStemmer method from nltk library
lancaster = LancasterStemmer()
def make_root(txt):                      #function that turns a word into its corresponding root word
    txt = lancaster.stem(txt)
    return txt 
```

## Main processing

Cleanup URLS, convert to lower case, tokenize, put everything in a list, remove stopwords, count, pass to dataframe, and visualize.

```python
# Remove URLs
#
tweetsWithoutURL = [remove_url(tweet) for tweet in textFromTweets]

# Convert to lower case
#
tweetsTokens = [tweet.lower().split() for tweet in tweetsWithoutURL]

# .chain(*arg) is a Python iterator;
# see https://docs.python.org/2/library/itertools.html#itertools.chain
# Here, we direct the iterator's output to a list() constructor
#
allWordsFromTweets = list(itertools.chain(*tweetsTokens))

# remove stopwords
#
allWordsFromTweets = [word 
                       for word 
                       in allWordsFromTweets
                       if not word in stopWords] 


##new
#makes a list of root words from the list of all words
allWordsWithRoots = [make_root(word) for word in allWordsFromTweets]
##new

# Count the occurences of each word now that trivial terms (stopwords)
# have been removed
#
wordCount = collections.Counter(allWordsFromTweets)

#new#
#a word count specifically for the root words
wordCountWithRoot = collections.Counter(allWordsWithRoots)
#new#

# Do a bit of reporting
#

print('      Tweets brought in: ' + str(TWEETS_TO_BRING_IN))
print('Non-trivial terms found: ' + str(len(allWordsFromTweets)))

# from python list to pandas dataframe
# (Remember: TOP_N_TERMS is defined in the beginning of the notebook)
#
tweetsDF = pd.DataFrame(wordCount.most_common(TOP_N_TERMS),
                        columns=['words', 'count'])

# plotting, finally
#
fig, ax = plt.subplots(figsize=(8, 8))

# Plot horizontal bar graph
#
tweetsDF.sort_values(by='count').plot.barh(x='words',
                      y='count',
                      ax=ax,
                      color="purple")

ax.set_title("The " + str(TOP_N_TERMS) + " most common terms in tweets about " + SEARCH_TERM)

plt.show()

```

```python

###plots a graph for all the root words

tweetsWithRootDF = pd.DataFrame(wordCountWithRoot.most_common(TOP_N_TERMS),
                        columns=['words', 'count'])                           #makes a data frame with the root word count

fig, ax = plt.subplots(figsize=(8, 8))

tweetsWithRootDF.sort_values(by='count').plot.barh(x='words',       #plots the root word data frame in pink
                      y='count',
                      ax=ax,
                      color="pink")

ax.set_title("The " + str(TOP_N_TERMS) + " most common root terms in tweets about " + SEARCH_TERM)   #specified "root terms" in the title

plt.show()


```

## Preserving the collection of documents

In addition to analyzing the frequency of terms of the aggragated documents, it may be worth lookign at some statistical properties on a per-document basis and across the collection.

![title](img/tokenization-within.png)

### Definitions

#### Term Frequency

The number of times a term (word) appears in each document (or, in this case, tweet), among a collection of documents. 

$\text{termFrequency}(t,d)=\dfrac{\text{count of term}\ t\ \text{in document}\ d}{\text{total number of terms in}\ d}$

If the documents are short, as is the case with tweets, we expect a term to appear 1 or 2 times in each document. 

#### Document Frequency

This is what we have been computing so far: the number of times each term appears across the entire collection of documents (or tweets).

$\text{documentFrequency}(t) = \text{Number of occurences of}\ t\ \text{across all documents}$

#### Inverse Document Frequency (IDF)

This is an expression that reflects the rarity of a term across the collection of $N$ documents.

$\text{inverseDocumentFrequency}(t)=\dfrac{N}{\text{documentFrequency}(t)}$

We expect the denominator to be always greater than zero, but just to be safe, we add 1, and we also take the logarithm of the fraction to avoid very large numbers.

$\text{inverseDocumentFrequency}(t)=\log{\left(\dfrac{N}{\text{documentFrequency}(t)+1}\right)}$

```python
# Compute the Document Frequencies
#
documentFrequency = defaultdict(int)

for text in tweetsWithoutURL:
    for word in text.lower().split():
        if not word in stopWords:
            documentFrequency[word] += 1

#document frequency for root words
documentFrequencyRoot = defaultdict(int)

for word in allWordsWithRoots:
    if not word in stopWords:
        documentFrequencyRoot[word] += 1
```
